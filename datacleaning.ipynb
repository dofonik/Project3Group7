```python
# Import dependencies
import pandas as pd
import numpy as np

world_data_df = pd.read_excel('resources/world-data-2023.xlsx')
world_data_df.head()
```

```python
# columns we want to keep
world_data_clean = ["country_name","Density\n(P/Km2)","Birth Rate","Co2-Emissions","CPI Change (%)","Fertility Rate", "GDP","Infant mortality","Life expectancy","Maternal mortality ratio","Minimum wage","Out of pocket health expenditure","Physicians per thousand","Population","Tax revenue (%)","Total tax rate","Unemployment rate","Urban_population","Latitude","Longitude"]
df_worldclean_data = world_data_df[world_data_clean]
df_worldclean_data.head()
```

```python
df_worldclean_data.dtypes
```

```python
df_worldclean_data.to_csv("output/worldcleandata.csv", index=False)
```
